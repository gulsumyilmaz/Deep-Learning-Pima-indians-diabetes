```python
#Normalizing is very important here, we are trying to bring the range 
#between 1 and 0 by dividing with 255.0. We added a 0 at the end so that 255.0 becomes a float
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
#CNN - Convolutional Neural Network 
#The following code provides us the ability to read and understand images
#LSTM long short term memory
```

```python
model=Sequential()
model.add(tf.keras.layers.InputLayer(input_shape=(28,28)))
model.add(tf.keras.layers.Reshape(target_shape=(28,28,1)))
model.add(tf.keras.layers.Conv2D(filters=12,kernel_size=(3,3),activation='relu'))
#The kernel matrix is taken as a 3x3 matrix, thus the image is also taken as such.
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
#Let's reduce it to a single dimension.
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(10))
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])  
```

```python
model.fit(train_images, train_labels, validation_split=0.10, epochs=10, verbose=1)  
```

```python
scores=model.evaluate(test_images,test_labels)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
model=Sequential([
    Flatten(input_shape=(28,28)),
    #rows turn into columns, to convert from multidimensional to one dimensional
    Dense(128,activation='relu'), #128 nöron olsun
    Dense(128,activation='relu'),
    Dense(128,activation='relu'),
    Dense(10)
])
```

```python
model.compile(optimizer="adam"
              ,loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=["accuracy"]
             )
```

```python
model.fit(train_images,train_labels,epochs=10,verbose=0)
#Images must be square. If verbose is set to 1, it shows what it's doing.
```
